```python
from langchain_community.document_loaders import WebBaseLoader
loader = WebBaseLoader("https://docs.langchain.com/oss/javascript/integrations/document_loaders/web_loaders")
documents = loader.load()

from langchain_text_splitters import CharacterTextSplitter

text_splitter = CharacterTextSplitter(chunk_size = 500, chunk_overlap=20, separator="\n")
chunks = text_splitter.split_documents(documents)
print(len(chunks[0].page_content))
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
```

```python


embeddings_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

chunk_embeddings = embeddings_model.embed_documents(
    [chunk.page_content for chunk in chunks]
)

print(len(chunk_embeddings[0]))
```
